Like Merge Sort, QuickSort is a Divide and Conquer algorithm. It picks an element as pivot and partitions the given array around the picked pivot.
There are many different versions of quickSort that pick pivot in different ways.
1. Always pick first element as pivot.
2. Always pick last element as pivot (implemented below)
3. Pick a random element as pivot.
4. Pick median as pivot.

```python
class Sorting_1:
    def quickSort(self, A, low, high):
        if A == None or len(A) < 1:
            return None
        if len(A) == 1:
            return A
        
        if low < high:
            pivot_index = self.partition(A, low, high)
            print('pivot_index', pivot_index)
            print('low', low)
            print('high', high)
            print('Enter Left half')
            self.quickSort(A, low, pivot_index-1)
            print('Enter right half')
            self.quickSort(A, pivot_index+1, high)
            print(A)
            print('quick sort over')
    
    def partition(self, A, low, high):
        i = low - 1
        pivot = A[high]
        print('pivot:', pivot)
        
        for j in range(low, high):
            if A[j] <= pivot:
                i += 1
                A[j], A[i] = A[i], A[j]
            print('i:', i, 'j:', j)
            print('A[i]:', A[i], 'A[j]', A[j])
            print(A)
        
        A[high], A[i+1] = A[i+1], A[high]
        print(A)
        print('partition over')
        return i+1
```

```python
A = [3,2,7,19,11,3,9,21]

s1 = Sorting_1()
s1.quickSort(A, 0, len(A)-1)
print(A)
```
